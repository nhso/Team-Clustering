```python
import pandas as pd
from sklearn.cluster import KMeans

data = pd.read_csv('Format.csv')

data.info
```

```python
# Select the columns you want to use for clustering
columns = ['1.1 ทะเบียนผู้มีสิทธิ', '1.2 ทะเบียนหน่วยบริการ', '1.3 บริการผู้ป่วยนอก (OP)', '1.4 บริการผู้ป่วยใน (IP)', '1.5 บริการสร้างเสริมสุขภาพและป้องกันโรค (PP)', '1.6 บริการอื่นๆ ที่เป็นลักษณะของการจัดการโรคเฉพาะ (DMIS)', '1.7 บริการฟื้นฟูสมรรถภาพฯ (Rehabilitation)', '1.8 งานบริการประชาชนและคุ้มครองสิทธิ', '1.9 การจัดสรรงบเหมาจ่ายรายหัว (Capitation)', '1.10 การเบิกจ่ายชดเชยค่าบริการ (Claims)', '1.11 การตรวจสอบการจ่ายชดเชย (Audit)', '1.12 ตัวชี้วัดตามเกณฑ์คุณภาพและผลงานบริการปฐมภูมิ', '1.13 บริการควบคุม ป้องกัน และรักษาโรคเรื้อรัง', '1.14 ตัวชี้วัดบริการสุขภาพ สปสช. (NHSO Health Service Indicator)', '1.15 รหัสมาตรฐานต่างๆที่เกี่ยวข้อง', '2.1 Descriptive Statistics', '2.2 Distribution', '2.3 Probability Theory', '2.4 Hypothesis Testing', '2.5 Correlation', '2.6 Regression', '2.7 Analysis of Variance: ANOVA)', '2.8 Discrete mathematics', '2.9 Boolean Algebra', '2.10 Math&Stat Tools', '3.1 SQL', '3.2 Data Manipulation / Transformation', '3.3 PL/SQL', '3.4 Data Pipelining', '3.5 Exploratory Data Analysis', '3.6 Prediction of Data', '3.7 Machine Learning', '3.8 Data Visualization', '3.9 Programming', '3.10 Networking ', '3.11 Software Architecture Design ', '3.12 Cyber Security', '3.13 Big Data Architecture', '3.14 Data Science Tool (R/Python)', '3.15 Transaction Processing System']

# Extract the data from the selected columns
X = data[columns]

# Specify the number of clusters you want to create
num_clusters = 7

# Create a KMeans instance and fit the data
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

# Get the cluster labels for each data point
labels = kmeans.labels_

# Add the cluster labels to the DataFrame
data['cluster'] = labels+1

# Print the updated DataFrame
data
```

```python
data.to_csv('output.csv', index=False)
```
