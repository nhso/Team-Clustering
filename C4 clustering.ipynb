```python
data_only_DDS = data.loc[data.division.isin(['ฝ่ายสนับสนุนข้อมูลดิจิทัล'])]
data_only_DDS
```

```python
merged = pd.merge(data, data_only_DDS, how='left', indicator=True)
data_without_DDS = merged[merged['_merge'] == 'left_only']
data_without_DDS = data_without_DDS.drop('_merge', axis=1)
data_without_DDS
```
