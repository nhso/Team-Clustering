```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np
```

```python
data_only_DDS = data.loc[data.division.isin(['ฝ่ายสนับสนุนข้อมูลดิจิทัล'])]
data_only_DDS
```

```python
merged = pd.merge(data, data_only_DDS, how='left', indicator=True)
data_no_DDS = merged[merged['_merge'] == 'left_only']
data_no_DDS = data_no_DDS.drop('_merge', axis=1)
data_no_DDS
```

```python
Data_finis = pd.concat([Spec1,Spec2,Spec3,data_rest,data_only_DDS])
Data_finis.reset_index(drop=True, inplace=True)
Data_finis = Data_finis.drop(['Mean','Rank','Mean1', 'Rank1','Mean2', 'Rank2','Mean3', 'Rank3'], axis=1)
Data_finis
```

```python
team_pattern = [1, 2, 3, 4, 5, 6, 7, 7, 6, 5, 4, 3, 2, 1]
teams = {}

for i, index in enumerate(Data_finis.index):
    team_num = team_pattern[i % len(team_pattern)]
    if team_num not in teams:
        teams[team_num] = []
    teams[team_num].append(Data_finis.loc[index])

for team_num, team_data in teams.items():
    team_df = pd.DataFrame(team_data)
    team_name = f"team{team_num}"
    globals()[team_name] = team_df

team1['Team'] = 1
team2['Team'] = 2
team3['Team'] = 3
team4['Team'] = 4
team5['Team'] = 5
team6['Team'] = 6
team7['Team'] = 7

C4_team = pd.concat([team1, team2, team3, team4, team5, team6, team7])
C4_team.reset_index(drop=True, inplace=True)

C4_team.to_csv('C4_team.csv')
```

```python
# Select the attributes for radar chart
attributes = ['Area1', 'Area2', 'Area3']

# Create traces for each person
traces = []
for _, row in C4_team.iterrows():
    values = row[attributes].tolist()
    trace = go.Scatterpolar(
        r=values,
        theta=attributes,
        fill='toself',
        name=row['ชื่อ-สกุล']
    )
    traces.append(trace)

# Create the radar chart
fig = go.Figure(data=traces)

# Set layout options
fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 10]  # Set the range of the radial axis
        )
    )
)

# Display the chart
fig.show()
# Create traces for each person
traces = []
for _, row in Team_B_Team.iterrows():
    values = row[attributes].tolist()
    trace = go.Scatterpolar(
        r=values,
        theta=attributes,
        fill='toself',
        name=row['Team']
    )
    traces.append(trace)

# Create the radar chart
fig = go.Figure(data=traces)

# Set layout options
fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 10]  # Set the range of the radial axis
        )
    )
)

# Display the chart
fig.show()
```
