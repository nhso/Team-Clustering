```python
for x in team1,team2,team3,team4,team5,team6,team7 :

# Normalize the attribute values to range from 0 to 1
  team_norm = (x[['Area1', 'Area2', 'Area3']] - x[['Area1', 'Area2', 'Area3']].min()) / (x[['Area1', 'Area2', 'Area3']].max() - x[['Area1', 'Area2', 'Area3']].min())

# Create a figure and subplot
  fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'polar': True})

# Calculate the angles for each attribute
  attributes = list(team_norm.columns)
  angles = [n / float(len(attributes)) * 2 * 3.14159 for n in range(len(attributes))]
  angles += angles[:1]

# Plot the spider chart for each person
  for i, row in team_norm.iterrows():
      values = row.values.tolist()
      values += values[:1]
      ax.plot(angles, values, linewidth=2, linestyle='solid', label=x['รหัสจนท.5หลัก'][i])

# Set the labels for each attribute
  ax.set_xticks(angles[:-1])
  ax.set_xticklabels(attributes)

# Add a legend
  ax.legend(loc='upper right', fontsize='small')

# Show the plot
plt.show()
```
